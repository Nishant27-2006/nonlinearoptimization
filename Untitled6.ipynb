```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, Bounds, NonlinearConstraint
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel as C
from collections import deque

class DynamicConstrainedProblem:
    def __init__(self, n_dim=5, l1_penalty=0.01, noise_level=0.005):
        self.n_dim = n_dim
        self.l1_penalty = l1_penalty
        self.noise_level = noise_level
        np.random.seed(42)
        self.obj_k = np.random.rand(n_dim) * 2 + 0.5
        self.obj_c = np.random.rand(n_dim) * 1.5
        self.theta_dim = 2
        self.true_theta_0 = np.array([0.5, 0.5])
        self.dyn_A = np.array([[0.1, 0.05], [0.03, 0.12]])
        self.dyn_B = np.array([[1.5, 0.8], [0.7, 1.2]])
        self.dyn_C = np.array([0.5, 0.3])
        self.dyn_D = np.array([[0.95, 0.01], [0.02, 0.93]])
        self.P_SMAX = {'A': 5, 'B': 3, 'C': 2 * np.pi}
        self.P_ALPHAS = {'E': 1.0, 'F': 0.5, 'G': np.pi, 'I': self.n_dim, 'H_const': 0.1}
        self.P_CBGT = {'A': 10, 'B': 4, 'C': 1.5 * np.pi}
        self.P_GAMMAS = {'E': 1.5, 'F': 0.7, 'G': np.pi, 'I': self.n_dim, 'J_const': 0.2}

    def objective_function(self, x):
        quad_terms = self.obj_k * x**2
        linear_terms = self.obj_c * x
        l1_term = self.l1_penalty * np.sum(np.abs(x))
        return np.sum(quad_terms - linear_terms) + l1_term

    def get_true_constraint_params(self, theta_k):
        theta_s, theta_c = theta_k[0], theta_k[1]
        s_max = self.P_SMAX['A'] + self.P_SMAX['B'] * np.sin(self.P_SMAX['C'] * theta_s)**2 + self.n_dim
        alphas = (self.P_ALPHAS['E'] + self.P_ALPHAS['F'] * np.cos(self.P_ALPHAS['G'] * theta_s + np.arange(self.n_dim) * self.P_ALPHAS['H_const'])) / self.P_ALPHAS['I']
        c_budget = self.P_CBGT['A'] + self.P_CBGT['B'] * np.sin(self.P_CBGT['C'] * theta_c)**2 + self.n_dim
        gammas = (self.P_GAMMAS['E'] + self.P_GAMMAS['F'] * np.sin(self.P_GAMMAS['G'] * theta_c + np.arange(self.n_dim) * self.P_GAMMAS['J_const'])) / self.P_GAMMAS['I']
        return alphas, s_max, gammas, c_budget

    def get_constraint_param_derivatives_wrt_theta(self, theta_k):
        theta_s, theta_c = theta_k[0], theta_k[1]
        ds_max_dtheta_s = self.P_SMAX['B'] * self.P_SMAX['C'] * np.sin(2 * self.P_SMAX['C'] * theta_s)
        H_i_array = np.arange(self.n_dim) * self.P_ALPHAS['H_const']
        dalphas_dtheta_s = -self.P_ALPHAS['F'] * self.P_ALPHAS['G'] * np.sin(self.P_ALPHAS['G'] * theta_s + H_i_array) / self.P_ALPHAS['I']
        dc_budget_dtheta_c = self.P_CBGT['B'] * self.P_CBGT['C'] * np.sin(2 * self.P_CBGT['C'] * theta_c)
        J_i_array = np.arange(self.n_dim) * self.P_GAMMAS['J_const']
        dgammas_dtheta_c = self.P_GAMMAS['F'] * self.P_GAMMAS['G'] * np.cos(self.P_GAMMAS['G'] * theta_c + J_i_array) / self.P_GAMMAS['I']
        return ds_max_dtheta_s, dalphas_dtheta_s, dc_budget_dtheta_c, dgammas_dtheta_c

    def true_constraints(self, x, theta_k):
        alphas, s_max, gammas, c_budget = self.get_true_constraint_params(theta_k)
        g1 = np.sum(alphas * x**2) - s_max
        g2 = np.sum(gammas * x) - c_budget
        return np.array([g1, g2])

    def update_true_theta(self, theta_k_current, x_k_chosen):
        mean_x_effect = np.mean(x_k_chosen) if x_k_chosen is not None else 0
        term_in_sin = self.dyn_B @ theta_k_current + self.dyn_C * mean_x_effect
        theta_k_next = self.dyn_A @ np.sin(term_in_sin) + self.dyn_D @ theta_k_current + np.random.normal(0, self.noise_level, self.theta_dim)
        return np.clip(theta_k_next, 0.0, 1.0)

class OnlineDynamicsLearnerGP:
    def __init__(self, theta_dim, n_dim, window_size=30, gp_retrain_interval=5):
        self.theta_dim = theta_dim; self.n_dim = n_dim; self.window_size = window_size
        self.gp_retrain_interval = gp_retrain_interval; self.steps_since_last_retrain = 0
        self.feature_dim = self.theta_dim + 1
        kernel = C(1.0, (1e-3, 1e3)) * RBF(length_scale=[1.0]*self.feature_dim, length_scale_bounds=(1e-2, 1e2)) + WhiteKernel(noise_level=0.01, noise_level_bounds=(1e-5, 1e1))
        self.gp_s = GaussianProcessRegressor(kernel=kernel, alpha=1e-7, n_restarts_optimizer=5, normalize_y=True)
        self.gp_c = GaussianProcessRegressor(kernel=kernel, alpha=1e-7, n_restarts_optimizer=5, normalize_y=True)
        self.history_X = deque(maxlen=self.window_size); self.history_y_s = deque(maxlen=self.window_size); self.history_y_c = deque(maxlen=self.window_size)
        self.last_predicted_theta_mean = None; self.last_predicted_theta_std = None; self.is_fitted = False
    def _construct_features(self, theta_k, x_k_mean): return np.concatenate([np.asarray(theta_k).flatten(), [x_k_mean]])
    def initialize(self, initial_theta_guess): self.last_predicted_theta_mean = np.copy(initial_theta_guess); self.last_predicted_theta_std = np.ones_like(initial_theta_guess) * 0.5
    def get_prediction_for_optimizer(self): return self.last_predicted_theta_mean
    def get_last_prediction_uncertainty(self): return self.last_predicted_theta_std
    def update_model_and_predict_for_next_step(self, theta_k_true, x_k_chosen, theta_k_plus_1_true):
        x_k_mean = np.mean(x_k_chosen) if x_k_chosen is not None else 0; phi_k = self._construct_features(theta_k_true, x_k_mean)
        self.history_X.append(phi_k); self.history_y_s.append(theta_k_plus_1_true[0]); self.history_y_c.append(theta_k_plus_1_true[1]); self.steps_since_last_retrain += 1
        if len(self.history_X) > self.feature_dim * 2 and self.steps_since_last_retrain >= self.gp_retrain_interval:
            current_X = np.array(self.history_X); current_y_s = np.array(self.history_y_s); current_y_c = np.array(self.history_y_c)
            try: self.gp_s.fit(current_X, current_y_s); self.gp_c.fit(current_X, current_y_c); self.is_fitted = True; self.steps_since_last_retrain = 0
            except Exception: self.is_fitted = False
        phi_for_next_pred = self._construct_features(theta_k_plus_1_true, x_k_mean); phi_for_next_pred_reshaped = phi_for_next_pred.reshape(1, -1)
        if self.is_fitted:
            try:
                pred_mean_s, pred_std_s = self.gp_s.predict(phi_for_next_pred_reshaped, return_std=True); pred_mean_c, pred_std_c = self.gp_c.predict(phi_for_next_pred_reshaped, return_std=True)
                self.last_predicted_theta_mean = np.array([pred_mean_s[0], pred_mean_c[0]]); self.last_predicted_theta_std = np.array([pred_std_s[0], pred_std_c[0]])
            except Exception: self.last_predicted_theta_mean = np.clip(theta_k_plus_1_true + np.random.normal(0,0.1,self.theta_dim), 0, 1); self.last_predicted_theta_std = np.ones_like(self.last_predicted_theta_mean) * 0.5
        else: self.last_predicted_theta_mean = np.clip(theta_k_plus_1_true + np.random.normal(0,0.1,self.theta_dim),0,1); self.last_predicted_theta_std = np.ones_like(self.last_predicted_theta_mean) * 0.5
        self.last_predicted_theta_mean = np.clip(self.last_predicted_theta_mean, 0.0, 1.0); self.last_predicted_theta_std = np.clip(self.last_predicted_theta_std, 1e-4, 1.0)

class AdaptiveOptimizerNovel:
    def __init__(self, problem_definition, dynamics_learner,
                 initial_kappa=1.5, min_kappa=0.5, max_kappa=3.0,
                 kappa_adapt_window=5, kappa_inc_factor=1.2, kappa_dec_factor=0.9,
                 objective_uncertainty_penalty_factor=0.05, adaptive_kappa_enabled=True):
        self.problem = problem_definition
        self.learner = dynamics_learner
        self.x_current = np.random.rand(self.problem.n_dim) * 2 - 1
        self.current_kappa = initial_kappa
        self.min_kappa = min_kappa; self.max_kappa = max_kappa
        self.kappa_adapt_window = kappa_adapt_window
        self.kappa_inc_factor = kappa_inc_factor; self.kappa_dec_factor = kappa_dec_factor
        self.lambda_uncertainty_penalty = objective_uncertainty_penalty_factor
        self.adaptive_kappa_enabled = adaptive_kappa_enabled

    def adapt_kappa(self, recent_true_g1_violations, recent_true_g2_violations):
        if not self.adaptive_kappa_enabled:
            return
        if not recent_true_g1_violations and not recent_true_g2_violations: return
        num_violations = sum(recent_true_g1_violations) + sum(recent_true_g2_violations)
        if num_violations > 0 : self.current_kappa = min(self.max_kappa, self.current_kappa * self.kappa_inc_factor)
        else: self.current_kappa = max(self.min_kappa, self.current_kappa * self.kappa_dec_factor)

    def optimize_step(self, k_iter, recent_true_g1_violations, recent_true_g2_violations):
        if k_iter >= self.kappa_adapt_window : self.adapt_kappa(recent_true_g1_violations, recent_true_g2_violations)
        predicted_theta_k_mean = self.learner.get_prediction_for_optimizer()
        predicted_theta_k_std = self.learner.get_last_prediction_uncertainty()
        if predicted_theta_k_mean is None:
             predicted_theta_k_mean = self.problem.true_theta_0
             predicted_theta_k_std = np.ones_like(predicted_theta_k_mean) * 0.5
        original_obj_fun = self.problem.objective_function
        alphas_mean, s_max_mean, gammas_mean, c_budget_mean = self.problem.get_true_constraint_params(predicted_theta_k_mean)
        ds_max_dtheta_s, dalphas_dtheta_s, dc_budget_dtheta_c, dgammas_dtheta_c = self.problem.get_constraint_param_derivatives_wrt_theta(predicted_theta_k_mean)
        pred_std_s = predicted_theta_k_std[0]; pred_std_c = predicted_theta_k_std[1]

        def get_constraint_uncertainties(x):
            dg1_dtheta_s_val = np.sum(dalphas_dtheta_s * x**2) - ds_max_dtheta_s
            sigma_g1 = np.abs(dg1_dtheta_s_val) * pred_std_s
            dg2_dtheta_c_val = np.sum(dgammas_dtheta_c * x) - dc_budget_dtheta_c
            sigma_g2 = np.abs(dg2_dtheta_c_val) * pred_std_c
            return sigma_g1, sigma_g2
        def uncertainty_penalized_objective(x):
            base_obj = original_obj_fun(x)
            sigma_g1_at_x, sigma_g2_at_x = get_constraint_uncertainties(x)
            penalty = self.lambda_uncertainty_penalty * (sigma_g1_at_x + sigma_g2_at_x)
            return base_obj + penalty
        def cons_g1(x):
            g1_mean = np.sum(alphas_mean * x**2) - s_max_mean
            sigma_g1_at_x, _ = get_constraint_uncertainties(x)
            return -(g1_mean + self.current_kappa * sigma_g1_at_x)
        def cons_g2(x):
            g2_mean = np.sum(gammas_mean * x) - c_budget_mean
            _, sigma_g2_at_x = get_constraint_uncertainties(x)
            return -(g2_mean + self.current_kappa * sigma_g2_at_x)

        constraints = [{'type': 'ineq', 'fun': cons_g1}, {'type': 'ineq', 'fun': cons_g2}]
        bounds = Bounds([-10]*self.problem.n_dim, [10]*self.problem.n_dim)
        options = {'maxiter': 50, 'disp': False}
        sol = minimize(uncertainty_penalized_objective, self.x_current, method='SLSQP', bounds=bounds, constraints=constraints, options=options)

        actual_x_chosen = self.x_current
        if sol.success: actual_x_chosen = sol.x
        self.x_current = actual_x_chosen
        obj_val_original = original_obj_fun(self.x_current)
        grad_norm = np.linalg.norm(sol.jac) if hasattr(sol, 'jac') and sol.jac is not None else float('nan')
        mean_g1_at_xk = np.sum(alphas_mean * self.x_current**2) - s_max_mean
        sigma_g1_at_xk, sigma_g2_at_xk = get_constraint_uncertainties(self.x_current)
        safety_margin_g1 = self.current_kappa * sigma_g1_at_xk
        mean_g2_at_xk = np.sum(gammas_mean * self.x_current) - c_budget_mean
        safety_margin_g2 = self.current_kappa * sigma_g2_at_xk
        targeted_g1_conservative = mean_g1_at_xk + safety_margin_g1
        targeted_g2_conservative = mean_g2_at_xk + safety_margin_g2
        return self.x_current, obj_val_original, grad_norm, np.array([mean_g1_at_xk, mean_g2_at_xk]), np.array([safety_margin_g1, safety_margin_g2]), np.array([targeted_g1_conservative, targeted_g2_conservative]), self.current_kappa

def run_simulation(config):
    problem = DynamicConstrainedProblem(
        n_dim=config['n_dim'],
        l1_penalty=config.get('l1_penalty', 0.01),
        noise_level=config['noise_level']
    )
    learner = OnlineDynamicsLearnerGP(
        problem.theta_dim, problem.n_dim,
        window_size=config.get('gp_window_size', 40),
        gp_retrain_interval=config.get('gp_retrain_interval', 3)
    )
    optimizer = AdaptiveOptimizerNovel(
        problem, learner,
        initial_kappa=config['initial_kappa'],
        min_kappa=config.get('min_kappa', 0.5),
        max_kappa=config.get('max_kappa', 3.0),
        kappa_adapt_window=config.get('kappa_adapt_window', 5),
        kappa_inc_factor=config.get('kappa_inc_factor', 1.2),
        kappa_dec_factor=config.get('kappa_dec_factor', 0.9),
        objective_uncertainty_penalty_factor=config['lambda_uncertainty_penalty'],
        adaptive_kappa_enabled=config['adaptive_kappa_enabled']
    )
    history = {
        'iterations': [], 'objective_values': [], 'gradient_norms': [], 'true_theta_s': [], 'true_theta_c': [],
        'predicted_theta_s_mean': [], 'predicted_theta_c_mean': [], 'predicted_theta_s_std': [], 'predicted_theta_c_std': [],
        'x_norm': [], 'true_g1_values': [], 'true_g2_values': [], 'mean_predicted_g1_values': [], 'mean_predicted_g2_values': [],
        'safety_margin_g1': [], 'safety_margin_g2': [], 'targeted_g1_conservative': [], 'targeted_g2_conservative': [], 'adaptive_kappa': []
    }
    recent_true_g1_violations_window = deque(maxlen=optimizer.kappa_adapt_window)
    recent_true_g2_violations_window = deque(maxlen=optimizer.kappa_adapt_window)
    current_true_theta = np.copy(problem.true_theta_0)
    initial_theta_guess_for_opt = np.copy(problem.true_theta_0)
    learner.initialize(np.clip(initial_theta_guess_for_opt,0,1))

    n_iterations = config['n_iterations']
    for k in range(n_iterations):
        pred_theta_for_opt_step_k_mean = learner.get_prediction_for_optimizer()
        pred_theta_for_opt_step_k_std = learner.get_last_prediction_uncertainty()
        x_k_chosen, obj_val_k, grad_norm_k, mean_pred_constraints_k, safety_margins_k, targeted_conservative_constraints_k, current_kappa_val = optimizer.optimize_step(k, list(recent_true_g1_violations_window), list(recent_true_g2_violations_window))
        next_true_theta = problem.update_true_theta(current_true_theta, x_k_chosen)
        learner.update_model_and_predict_for_next_step(current_true_theta, x_k_chosen, next_true_theta)
        history['iterations'].append(k); history['objective_values'].append(obj_val_k); history['gradient_norms'].append(grad_norm_k)
        history['true_theta_s'].append(current_true_theta[0]); history['true_theta_c'].append(current_true_theta[1])
        history['predicted_theta_s_mean'].append(pred_theta_for_opt_step_k_mean[0]); history['predicted_theta_c_mean'].append(pred_theta_for_opt_step_k_mean[1])
        history['predicted_theta_s_std'].append(pred_theta_for_opt_step_k_std[0]); history['predicted_theta_c_std'].append(pred_theta_for_opt_step_k_std[1])
        history['x_norm'].append(np.linalg.norm(x_k_chosen))
        true_constraints_vals_k = problem.true_constraints(x_k_chosen, current_true_theta)
        history['true_g1_values'].append(true_constraints_vals_k[0]); history['true_g2_values'].append(true_constraints_vals_k[1])
        recent_true_g1_violations_window.append(true_constraints_vals_k[0] > 1e-3)
        recent_true_g2_violations_window.append(true_constraints_vals_k[1] > 1e-3)
        history['mean_predicted_g1_values'].append(mean_pred_constraints_k[0]); history['mean_predicted_g2_values'].append(mean_pred_constraints_k[1])
        history['safety_margin_g1'].append(safety_margins_k[0]); history['safety_margin_g2'].append(safety_margins_k[1])
        history['targeted_g1_conservative'].append(targeted_conservative_constraints_k[0]); history['targeted_g2_conservative'].append(targeted_conservative_constraints_k[1])
        history['adaptive_kappa'].append(current_kappa_val)
        current_true_theta = next_true_theta
        if k % (n_iterations // 5) == 0 or k == n_iterations - 1 :
             if k != 0 : print(f"  Iter {k}: Obj={obj_val_k:.2f}, Kappa={current_kappa_val:.2f}, True g1={true_constraints_vals_k[0]:.2f}")
    return history, problem

def plot_results(history, problem, config_name, config_params):
    iterations = history['iterations']

    fig1, axs1 = plt.subplots(2, 1, figsize=(10, 8))
    fig1.suptitle(f"Optimizer Performance ({config_name})\n(n_dim={config_params['n_dim']}, init_kappa={config_params['initial_kappa']}, obj_penalty={config_params['lambda_uncertainty_penalty']}, noise={config_params['noise_level']})", fontsize=12)
    axs1[0].plot(iterations, history['objective_values'], marker='o', markersize=3, linestyle='-', label="Objective Value (Original)")
    axs1[0].set_xlabel("Iteration"); axs1[0].set_ylabel("Objective Value"); axs1[0].set_title("Objective Function Evolution")
    axs1[0].grid(True); axs1[0].legend()
    axs1[1].plot(iterations, history['gradient_norms'], marker='o', markersize=3, linestyle='-', color='r', label="Gradient Norm (Penalized Obj)")
    axs1[1].set_xlabel("Iteration"); axs1[1].set_ylabel("Gradient Norm"); axs1[1].set_yscale('log')
    axs1[1].set_title("Gradient Norm Evolution"); axs1[1].grid(True); axs1[1].legend()
    plt.tight_layout(rect=[0, 0, 1, 0.93])
    plt.show()

    fig2, axs2 = plt.subplots(2, 1, figsize=(10, 8))
    fig2.suptitle(f"Learning of $\\theta_s$ ({config_name})", fontsize=14)
    true_s = np.array(history['true_theta_s']); pred_s_mean = np.array(history['predicted_theta_s_mean']); pred_s_std = np.array(history['predicted_theta_s_std'])
    axs2[0].plot(iterations, true_s, 'k-', marker='x', markersize=4, label="True $\\theta_s$"); axs2[0].plot(iterations, pred_s_mean, 'b-', marker='.', markersize=4, label="GP Mean $\\hat{\\theta}_s$")
    axs2[0].fill_between(iterations, pred_s_mean - 1.96 * pred_s_std, pred_s_mean + 1.96 * pred_s_std, alpha=0.2, color='blue', label="GP 95% CI $\\hat{\\theta}_s$")
    axs2[0].set_xlabel("Iteration"); axs2[0].set_ylabel("$\\theta_s$ Value"); axs2[0].set_title("Evolution with GP Uncertainty"); axs2[0].legend(); axs2[0].grid(True)
    axs2[1].plot(iterations, pred_s_std, marker='.', markersize=4, linestyle='-', color='blue', label="GP Pred Std Dev $\\hat{\\theta}_s$")
    axs2[1].set_xlabel("Iteration"); axs2[1].set_ylabel("Std Dev"); axs2[1].set_title("GP Prediction Uncertainty"); axs2[1].grid(True); axs2[1].legend(); axs2[1].set_yscale('log')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    fig3, axs3 = plt.subplots(2, 1, figsize=(10, 8))
    fig3.suptitle(f"Learning of $\\theta_c$ ({config_name})", fontsize=14)
    true_c = np.array(history['true_theta_c']); pred_c_mean = np.array(history['predicted_theta_c_mean']); pred_c_std = np.array(history['predicted_theta_c_std'])
    axs3[0].plot(iterations, true_c, 'k-', marker='x', markersize=4, label="True $\\theta_c$"); axs3[0].plot(iterations, pred_c_mean, 'g-', marker='.', markersize=4, label="GP Mean $\\hat{\\theta}_c$")
    axs3[0].fill_between(iterations, pred_c_mean - 1.96 * pred_c_std, pred_c_mean + 1.96 * pred_c_std, alpha=0.2, color='green', label="GP 95% CI $\\hat{\\theta}_c$")
    axs3[0].set_xlabel("Iteration"); axs3[0].set_ylabel("$\\theta_c$ Value"); axs3[0].set_title("Evolution with GP Uncertainty"); axs3[0].legend(); axs3[0].grid(True)
    axs3[1].plot(iterations, pred_c_std, marker='.', markersize=4, linestyle='-', color='green', label="GP Pred Std Dev $\\hat{\\theta}_c$")
    axs3[1].set_xlabel("Iteration"); axs3[1].set_ylabel("Std Dev"); axs3[1].set_title("GP Prediction Uncertainty"); axs3[1].grid(True); axs3[1].legend(); axs3[1].set_yscale('log')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    fig4, axs4 = plt.subplots(2, 1, figsize=(10, 8))
    fig4.suptitle(f"Analysis of Constraint $g_1$ ({config_name})", fontsize=14)
    axs4[0].plot(iterations, history['true_g1_values'], marker='s', markersize=3, linestyle='-', color='black', label="True $g_1(x_k, \\theta_k^{true})$")
    axs4[0].plot(iterations, history['mean_predicted_g1_values'], marker='^', markersize=3, linestyle=':', color='orange', label="Mean Pred $g_1(x_k, \\hat{\\theta}_k)$")
    axs4[0].plot(iterations, history['targeted_g1_conservative'], marker='+', markersize=4, linestyle='--', color='red', label="Targeted $g_1^{cons}(x_k)$ by Opt")
    axs4[0].axhline(0, color='grey', linestyle='--', linewidth=0.8, label="$g_1=0$ Boundary")
    axs4[0].set_xlabel("Iteration"); axs4[0].set_ylabel("Constraint $g_1$ Value"); axs4[0].set_title("Constraint Values"); axs4[0].legend(); axs4[0].grid(True)
    axs4[1].plot(iterations, history['safety_margin_g1'], marker='.', markersize=4, linestyle='-', color='red', label="Safety Margin for $g_1$ ($\kappa \sigma_{g1}$)")
    axs4[1].set_xlabel("Iteration"); axs4[1].set_ylabel("Margin Value"); axs4[1].set_title("Applied Safety Margin"); axs4[1].grid(True); axs4[1].legend()
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    fig5, axs5 = plt.subplots(2, 1, figsize=(10, 8))
    fig5.suptitle(f"Analysis of Constraint $g_2$ ({config_name})", fontsize=14)
    axs5[0].plot(iterations, history['true_g2_values'], marker='s', markersize=3, linestyle='-', color='black', label="True $g_2(x_k, \\theta_k^{true})$")
    axs5[0].plot(iterations, history['mean_predicted_g2_values'], marker='^', markersize=3, linestyle=':', color='limegreen', label="Mean Pred $g_2(x_k, \\hat{\\theta}_k)$")
    axs5[0].plot(iterations, history['targeted_g2_conservative'], marker='+', markersize=4, linestyle='--', color='purple', label="Targeted $g_2^{cons}(x_k)$ by Opt")
    axs5[0].axhline(0, color='grey', linestyle='--', linewidth=0.8, label="$g_2=0$ Boundary")
    axs5[0].set_xlabel("Iteration"); axs5[0].set_ylabel("Constraint $g_2$ Value"); axs5[0].set_title("Constraint Values"); axs5[0].legend(); axs5[0].grid(True)
    axs5[1].plot(iterations, history['safety_margin_g2'], marker='.', markersize=4, linestyle='-', color='purple', label="Safety Margin for $g_2$ ($\kappa \sigma_{g2}$)")
    axs5[1].set_xlabel("Iteration"); axs5[1].set_ylabel("Margin Value"); axs5[1].set_title("Applied Safety Margin"); axs5[1].grid(True); axs5[1].legend()
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    fig6, axs6 = plt.subplots(2, 2, figsize=(12, 8))
    fig6.suptitle(f"Solution Characteristics and Adaptive Kappa ({config_name})", fontsize=14)
    axs6[0,0].plot(iterations, history['x_norm'], marker='p', markersize=4, linestyle='-', color='magenta', label="$||x_k||_2$")
    axs6[0,0].set_xlabel("Iteration"); axs6[0,0].set_ylabel("Norm of $x_k$"); axs6[0,0].set_title("Norm of Decision Variables"); axs6[0,0].grid(True); axs6[0,0].legend()
    true_g1_violations = np.array(history['true_g1_values']) > 1e-3
    true_g2_violations = np.array(history['true_g2_values']) > 1e-3
    axs6[0,1].plot(iterations, true_g1_violations, marker='1', linestyle='none', color='red', markersize=8, label='True $g_1$ Violated')
    axs6[0,1].plot(iterations, true_g2_violations + 0.1 , marker='2', linestyle='none', color='maroon', markersize=8, label='True $g_2$ Violated')
    axs6[0,1].set_xlabel("Iteration"); axs6[0,1].set_ylabel("Violation (Binary)"); axs6[0,1].set_title("True Constraint Violations"); axs6[0,1].set_yticks([0, 1]); axs6[0,1].set_yticklabels(['Feasible', 'Violated']); axs6[0,1].legend(); axs6[0,1].grid(True)
    axs6[1,0].plot(iterations, history['adaptive_kappa'], marker='.', markersize=4, linestyle='-', color='teal', label="Adaptive $\kappa$")
    axs6[1,0].set_xlabel("Iteration"); axs6[1,0].set_ylabel("$\kappa$ Value"); axs6[1,0].set_title("Evolution of Adaptive Safety Factor $\kappa$"); axs6[1,0].grid(True); axs6[1,0].legend()
    if not config_params['adaptive_kappa_enabled']:
        axs6[1,0].axhline(config_params['initial_kappa'], color='grey', linestyle=':', label=f"Fixed $\kappa={config_params['initial_kappa']}$")
        axs6[1,0].legend()
    axs6[1,1].axis('off')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

if __name__ == "__main__":
    configurations = [
        {
            'name': "Default (Dim 5, Adapt Kappa, Low Penalty, Low Noise)",
            'n_iterations': 100, 'n_dim': 5, 'initial_kappa': 1.0,
            'lambda_uncertainty_penalty': 0.01, 'noise_level': 0.005,
            'adaptive_kappa_enabled': True
        },
        {
            'name': "High Dim (Dim 10, Adapt Kappa, Low Penalty, Low Noise)",
            'n_iterations': 100, 'n_dim': 10, 'initial_kappa': 1.0,
            'lambda_uncertainty_penalty': 0.01, 'noise_level': 0.005,
            'adaptive_kappa_enabled': True
        },
        {
            'name': "Fixed High Kappa (Dim 5, No Adapt, High Penalty, Low Noise)",
            'n_iterations': 100, 'n_dim': 5, 'initial_kappa': 2.5,
            'lambda_uncertainty_penalty': 0.05,
            'noise_level': 0.005,
            'adaptive_kappa_enabled': False
        },
        {
            'name': "High Noise (Dim 5, Adapt Kappa, Mid Penalty, High Noise)",
            'n_iterations': 100, 'n_dim': 5, 'initial_kappa': 1.5,
            'lambda_uncertainty_penalty': 0.03,
            'noise_level': 0.05,
            'adaptive_kappa_enabled': True
        },
         {
            'name': "No Objective Penalty (Dim 5, Adapt Kappa, No Penalty, Low Noise)",
            'n_iterations': 100, 'n_dim': 5, 'initial_kappa': 1.0,
            'lambda_uncertainty_penalty': 0.0,
            'noise_level': 0.005,
            'adaptive_kappa_enabled': True
        },
    ]

    for config in configurations:
        print(f"\n\n======================================================================")
        print(f"Running Simulation for Configuration: {config['name']}")
        print(f"Parameters: {config}")
        print(f"======================================================================")

        simulation_history, problem_instance = run_simulation(config)

        print(f"\n--- Plotting results for: {config['name']} ---")
        plot_results(simulation_history, problem_instance, config['name'], config)
        print(f"--- Plotting complete for: {config['name']} ---")

    print("\nAll configurations processed.")
```
